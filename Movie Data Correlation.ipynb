```python
#import libraries needed
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib import figure

#Magic function to allow graphs to be included in notebook
%matplotlib inline

#Configuring the plot to be used
matplotlib.rcParams['figure.figsize'] = (12,8)

#Read data into a dataframe
df = pd.read_csv(r'movies.csv')

```

```python
#View a snapshot of data
df.head()
```

```python
#Check for missing data
for col in df.columns:
    missing_pct = np.mean(df[col].isnull())
    print('{}  - {}%'.format(col, round(missing_pct*100,2)))
```

```python
#Data types of columns
df.dtypes
```

```python
#Cleaning out data set by dropping nan values
df = df.dropna()
```

```python
#Changing data type of Budget and Gross to an integer
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

#Confirm data type has been changed
df.head()
```

#Upon analyzing data, the year in the released column doesn't match the year in the year column
#To fix that, a new column has to be created by extracting the year from the released column

```python
df['correct_year'] = df['released'].astype(str).str.extract(pat = '([0-9]{4})').astype(int)

#View data
df.head()
```

```python
#Sort movies by highest to least grossing 
df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
#Set data output to show all entries
pd.set_option('display.max_rows', None)
```

```python
#Drop any duplicates to see all distinct companies

df['company'].drop_duplicates().sort_values(ascending = False).head()
```

```python
#High Budget Correlation
#Inferences
```

```python
#Scatter plot with budget vs gross

plt.scatter(x = df['budget'], y= df['gross'])
plt.title('Bugdet vs Gross Earnings')
plt.ylabel('Gross earnings')
plt.xlabel('Budget for Film')
plt.show()
```

```python
df.corr(method = "pearson") #pearson, kendall, spearman
```

```python
#Correlation using Pearson Method

correlation_matrix = df.corr(method = "pearson")
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix - Pearson')
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')

plt.show()
```

```python
#Correlation using Kendall Method

correlation_matrix = df.corr(method = "kendall")
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix - Kendall')
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')

plt.show()
```

```python
#Correlation using Spearman Method

correlation_matrix = df.corr(method = "spearman")
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix - Spearman')
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()        
```

```python
df.head()
```

```python
#Correlation using Pearson Method after numerizing data

correlation_matrix = df_numerized.corr(method = "pearson")
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix - Pearson')
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')

plt.show()
```

```python
correction_mat = df_numerized.corr()

corr_pairs = correction_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
